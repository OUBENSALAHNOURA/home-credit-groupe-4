```python
from tqdm import tqdm
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://admin:19961994@database-1.c6j2ix3ngibe.us-east-1.rds.amazonaws.com/homecreditprojet', echo=False)

def insert(df, l):
    dfs = spliter_df(df, l)
    first = True
    case = "replace"
    for df in tqdm(dfs):
        if not first:
            case = "append"
        insert_one(df, case)
```

```python
def spliter_df(df, l):
    size = len(df)
    nb = int(size /l)
    return [df[d:(1+d)*l] for d in range(nb)]
    
```

```python
def  insert_one(df, case):
    try:
        df.to_sql(name= 'bureau_balance' ,con=engine,index=False, if_exists=case)
    except Exception as e:
        print(e)
        return False
    return True
```
